```python
import pandas as pd
df = pd.read_csv('path_to_file',  sep=' ')
```

```python
print(df)
```

```python
import csv
col_names = df.columns.values
lenght = len(col_names)
i=1
while i < lenght:
    file_name = 'path_to_output_directory'
    export_col = [col_names[0]]
    export_name = col_names[i]
    export_col.append(export_name)
    trimmed_name = export_name.replace(':', '')
    file_name += trimmed_name
    file_name += '.csv'
    df.to_csv(file_name, columns=export_col, index=False,  sep=' ')
    print(export_col)
    i += 1
```
